```python
#dropping unused column
supervised_dataset_treated = supervised_dataset.drop(['Unnamed: 0'], axis=1)
```

```python
#knowing the ip_type categories
supervised_dataset_treated['ip_type'].unique()
```

```python
#replace categories to normalize ////// defaulf = -1  -------  datacenter = 1
supervised_dataset_treated['ip_type'] = supervised_dataset_treated['ip_type'].replace('default', -1).replace('datacenter', 1)
```

```python
#knowing the source categories
supervised_dataset_treated['source'].unique()
```

```python
#replace categories to normalize ////// E = -1  -------  F = 1
supervised_dataset_treated['source'] = supervised_dataset_treated['source'].replace('E', -1).replace('F', 1)
```

```python
#knowing the classification categories
supervised_dataset_treated['classification'].unique()
```

```python
#replace categories to normalize ////// E = -1  -------  F = 1
supervised_dataset_treated['classification'] = supervised_dataset_treated['classification'].replace('normal', 0).replace('outlier', 1)
```

```python
supervised_dataset_treated.to_csv('../dataset/normalized/supervised_dataset_treated')
```
